## Task 1: Reading and Analyzing the Data.
Before we start our analysis, first, we need to view the dataset. It is essential to view the data and check the columns. Let's take a look.

```python
import warnings
warnings.filterwarnings("ignore")

#--- Import Pandas ---
import pandas as pd
#--- Read in dataset ----
df = pd.read_csv('German_car_data.csv')
df
```

```python
df.head()
```

## Task 2: Dropping Column from DataFrame.
We have our dataset loaded. Now, we need to remove the unwanted columns. So, let's remove those columns.

```python
# Dropping the first column
#--- WRITE YOUR CODE FOR TASK 2 ---
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df.head()
```

## Task 3: Counting Null Values in the Data.
We have successfully remove the column. Now, we need to ensure that the dataset does not contain any null values. So, let's check for null values in our dataset. 

```python
null_values = df.isnull().sum()
null_values
```

## Task 4: Analyzing Missing Fuel Consumption Data by Fuel Type.
We've discovered null values in our dataset. Now, our next step is to ensure the dataset is free of any missing values. To do this, we're checking for null values specifically in the 'fuel_consumption_l_100km' column. Let's see how many missing values we have for each fuel type in our dataset. 

```python
missing_df1 = df.groupby('fuel_type')['fuel_consumption_l_100km'].apply(lambda x: x.isnull().sum()).reset_index(name='counts')

missing_df1 = missing_df1[missing_df1['counts'] > 0]

missing_df = missing_df1.sort_values(by = 'counts', ascending = False).set_index('fuel_type').squeeze()
missing_df
```

## Task 5: Handling Missing Fuel Consumption Data.
We've found null values in fuel_type. Now, we need to fill in these missing values to ensure our dataset is ready for further analysis. Let's proceed with filling them in. 

```python
# condition to identify rows where fuel consumption is null and the fuel type corresponds to electric hydrogen or ethanol
condition = df['fuel_consumption_l_100km'].isnull() & df['fuel_type'].isin(['Electric', 'Hydrogen', 'Ethanol'])

# assign 0 where fuel consumption is met the condition
df.loc[condition, 'fuel_consumption_l_100km'] = 0
```

## Task 6: Removing Unknown and Other Fuel Types.
We've successfully filled the values. Now, we need to remove the unwanted rows. By filtering out these entries, we streamline our dataset for clearer analysis. Let's take a look at the refined dataset without these entries. 

```python
# filtering out rows from the dataframe where fuel type is either unknown or other
filtered = df[~df['fuel_type'].isin(['Unknown', 'Other'])]

# Apply the condition to filter the DataFrame

filtered.shape
```

## Task 7: Removing Rows with Missing Values.
We've successfully removed the unwanted values. Now, we need to drop the rows containing any missing values. This helps maintain data quality and prepares our dataset for thorough analysis. Let's proceed with the cleaned dataset. 

```python
df = df.dropna()
```

## Task 8: Analyzing Car Listings by Year.
We've successfully removed the unwanted rows. Now, we need to examine the distribution of car listings by year.Let's explore it. 

```python
year_value = df['year'].value_counts()
```

## Task 9: Filtering Years in Dataset.
We've analyzed the years. Now, we need to create a list of recent years. By selecting only the year values, we aim for better data for analysis. Let's get to work on it. 

```python
yearsToFilter = df['year'].unique()[:29].tolist()
```

## Task 10: Filtering Dataset by Years.
We've successfully created a list of years. Now, we need to filtering our dataset to include only entries from year values. Applying this filter helps us focus our analysis on more cleaned data. Let's filter the dataset accordingly. 

```python
filtered_df = df[df['year'].isin(yearsToFilter)]
```

## Task 11: Extracting Fuel Consumption Range for Electric Cars.
We've successfully filtered the years. Now, we need to process the fuel_consumption_l_100km column. By doing this it helps us focus our analysis on more cleaned data. Let's process the dataset accordingly. 

```python
df['fuel_consumption_g_km'] = df.apply(
    lambda row: float(row['fuel_consumption_g_km'][:3]) 
    if 'Reichweite' in row['fuel_consumption_g_km'] and 
       row['fuel_type'] == 'Electric' and 
       'k' not in row['fuel_consumption_g_km'][:3] 
    else row['fuel_consumption_g_km'], axis=1
)
```

## Task 12: Standardizing Consumption Value Formatting.
We've successfully processed the data. Now, we need to process the fuel_consumption_l_100km column again for formatting. By doing this it helps us focus our analysis on more cleaned data. Let's process the dataset accordingly. 

```python
df['fuel_consumption_l_100km'] = df['fuel_consumption_l_100km'].apply(
    lambda x: float(str(x).replace(' kWh/100 km', '').replace(',', '.')) 
    if isinstance(x, str) and ' kWh/100 km' in x 
    else x
)
```

## Task 13: Standardizing Consumption Value Formatting.
We've successfully formatted the data. Now, we need to identify values in the 'fuel_consumption_l_100km' column that need to be dropped. By doing this step helps us identify and handle irrelevant data in our dataset. Let's mark these values for removal in the dataset. 

```python
df['fuel_consumption_l_100km'] = df.apply(
    lambda row: 'To Drop' 
    if (not isinstance(row['fuel_consumption_l_100km'], float) or row['fuel_consumption_l_100km'] == 0) 
       and row['fuel_type'] == 'Electric' 
    else row['fuel_consumption_l_100km'], axis=1
)
```

## Task 14: Removing Rows Marked for Dropping.
We've successfully processed the data. Now, we need to remove rows from the dataset that were marked in the last task. By filtering out these rows, we ensure the integrity of our dataset for further analysis. Let's proceed with removing these rows from the dataset. 

```python
filtered_df = df[df['fuel_consumption_l_100km'] != 'To Drop']
```

## Task 15: Standardizing Consumption Value Formatting for Liters per 100 km.
We've successfully dropped the unwanted entries. Now, we need to remove the ' l/100 km' suffix and replace any commas with periods to ensure consistent formatting. This step enhances the clarity and accuracy of our dataset for analysis. Let's implement this formatting adjustment to the dataset. 

```python
df['fuel_consumption_l_100km'] = df['fuel_consumption_l_100km'].apply(
    lambda x: float(x.replace(' l/100 km', '').replace(',', '.')) 
    if isinstance(x, str) and ' l/100 km' in x 
    else x
)
```

## Task 16: Standardizing Consumption Value Formatting for Kilograms per 100 km.
We've successfully dropped the suffix. Now, we need to remove the ' kg/100 km' suffix and replace any commas with periods to ensure consistent formatting. This step enhances the clarity and accuracy of our dataset for analysis. Let's implement this formatting adjustment to the dataset. 

```python
df['fuel_consumption_l_100km'] = df['fuel_consumption_l_100km'].apply(
    lambda x: float(x.replace(' kg/100 km', '').replace(',', '.')) 
    if isinstance(x, str) and ' kg/100 km' in x 
    else x
)
```

## Task 17: Filtering DataFrame for Float Values.
We've successfully dropped the suffix again. Now, we need to retain rows where the 'fuel_consumption_l_100km' column contains float values. By doing this, we ensure data integrity and prepare the dataset for further analysis. Let's take a look.

  

```python
# Ensure that the column 'fuel_consumption_l_100km' is of type float where applicable
df['fuel_consumption_l_100km'] = pd.to_numeric(df['fuel_consumption_l_100km'], errors='coerce')

# Filter rows where 'fuel_consumption_l_100km' is a float
filtered_df = df[df['fuel_consumption_l_100km'].notna()]
filtered_df
```

## Task 18: Converting Data Types.
We've successfully filtered the DataFrame. Now, we need to convert the data types of numbers to integer and floating point. By doing this, we ensure data integrity and prepare the dataset for further analysis. Let's apply these data type conversions to the DataFrame. 

```python
df['price_in_euro'] = pd.to_numeric(df['price_in_euro'], errors='coerce')

# Clean 'power_kw' column by removing non-numeric values and convert to numeric (use `errors='coerce'` to handle non-numeric values)
df['power_kw'] = pd.to_numeric(df['power_kw'], errors='coerce')

# Convert the 'price_in_euro' and 'power_kw' columns to integer, handling NaNs by filling them with a default value, e.g., 0
df['price_in_euro'] = df['price_in_euro'].fillna(0).astype(int)
df['power_kw'] = df['power_kw'].fillna(0).astype(int)

# Convert 'fuel_consumption_l_100km' column to float data type
df['fuel_consumption_l_100km'] = pd.to_numeric(df['fuel_consumption_l_100km'], errors='coerce')
df.dtypes
df
```

## Task 21: Visualizing Number of Car Listings by Brand.
We've successfully cleaned the dataframe for the analysis. Now, we're visualizing the distribution of car listings across different brands. By creating a horizontal bar plot, we showcase the number of listings for each brand in our dataset. This visualization provides insights into the popularity and presence of various car brands in the market. Let's explore the distribution of car listings per brand. 

```python
df.head(5)
```

```python
import matplotlib.pyplot as plt

# Count the number of listings for each brand
brand_counts = df['brand'].value_counts()

# Set the figure size to 10x6 inches
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
plot1 = plt.barh(brand_counts.index, brand_counts.values)

# Label the x-axis and y-axis
plt.xlabel('Number of Listings')
plt.ylabel('Brand')

# Set the title of the plot
plt.title('Number of Listings Per Brand')

# Display the plot
plt.show()
```

## Task 22: Analyzing Average Price Per Brand.
We've successfully plotted the distribution of car listings across different brands. Now, we're analyzing the average price of car listings for each brand in our dataset. This visualization helps us understand the relative pricing of cars from various manufacturers. Let's explore the average price per brand in our dataset. 

```python
brandPrice = df.groupby('brand')['price_in_euro'].mean()

# Set the figure size to 10x6 inches
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
plot2 = plt.barh(brandPrice.index, [brandPrice.loc[i] for i in brandPrice.index])

# Label the x-axis
plt.xlabel('Average Price (EUR)')

# Set the title of the plot
plt.title('Average Price Per Brand')

# Display the plot
plt.show()
```

##  Task 23: Analyzing Efficiency of Internal Combustion Engine (ICE) Cars Per Brand.
We've analyzed the average price of car listings for each brand. Now, we're analyzing the efficiency of internal combustion engine (ICE) cars for selected fuel types across different car brands. This visualization helps us understand the relative fuel efficiency of ICE cars across different brands in our dataset. Let's explore the efficiency of ICE cars per brand.

```python
df
```

```python
# Filter the DataFrame to include only rows where the 'fuel_type' is 'Petrol', 'Diesel', 'Hybrid', or 'Diesel Hybrid'
ice_df = df[df['fuel_type'].isin(['Petrol', 'Diesel', 'Hybrid', 'Diesel Hybrid',])]

# Compute the mean fuel consumption (liters per 100 km) for each brand
brandEff = ice_df.groupby('brand')['fuel_consumption_l_100km'].mean()
brandEff = brandEff.sort_values(ascending = False)

# Set the figure size to 10x6 inches
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
plot3 = plt.barh(brandEff.index, [brandEff.loc[i] for i in brandEff.index])

# Label the x-axis and y-axis
plt.xlabel('Litres of Fuel per 100km')
plt.ylabel('Brand')

# Set the title of the plot
plt.title('Efficiency of ICEs Per Brand')

# Display the plot
plt.show()
```

## Task 24: Analyzing Average Mileage Per Brand.
